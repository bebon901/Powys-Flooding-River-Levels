```python
#!pip install plotly
import urllib.request
from matplotlib import pyplot as plt
import numpy as np
import plotly.graph_objects as go
import pandas as pd
data = ""
for line in urllib.request.urlopen("https://bebon901.github.io/Powys-Flooding-River-Levels/river-levels-flood-2022.csv"):
    data += (line.decode('utf-8'))
content = data.split("\n")

time = []
llandrinio = []
llanymynech = []
abermule = []
caersws = []
meifod = []
index = []
llanidloes = []
c = 0
i = 1
d = ""
while (d != None):
    d = content[i].split(",")
    i+=1
    c+=0.25
    #print(d)
    if len(d) > 2:
        try:
            time.append(d[0])
            llandrinio.append(float(d[1]))
            index.append(c)
            llanymynech.append(float(d[2]))
            abermule.append(float(d[3]))
            meifod.append(float(d[5]))
            caersws.append(float(d[6]))
            llanidloes.append(float(d[7]))
        except:
            pass
    else:
        break
#print(llandrinio)
'''
plt.plot(index, llandrinio, label="Llandrinio")
#plt.plot(index, llanymynech, label="llanymynech")
plt.plot(index, abermule, label="Abermule")
#plt.plot(index, meifod, label="meifod")
plt.plot(index, caersws, label="Caersws")
plt.plot(index, llanidloes, label="Llanidloes")
plt.legend(loc="best")
plt.show()
'''


trace0 = go.Scatter(
   x = index, y = llandrinio, name='Llandrinio'
)
data = [trace0]
fig1 = go.Figure()

# Add a scatter plot trace for the measurements.
fig1.add_trace(trace0)



trace0 = go.Scatter(
   x = index, y = abermule, name='Abermule'
)
data = [trace0]
fig1.add_trace(trace0)

trace0 = go.Scatter(
   x = index, y = caersws, name='Caersws'
)
data = [trace0]
fig1.add_trace(trace0)

trace0 = go.Scatter(
   x = index, y = llanidloes, name='Llanidloes'
)
data = [trace0]
fig1.add_trace(trace0)

# Label the axes.
fig1.update_xaxes(title_text='Hours gone by...')
fig1.update_yaxes(title_text='Height (m)')

# Show the figure.
fig1.show()
```
